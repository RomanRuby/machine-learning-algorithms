BUILDING DECISION TREE(Scikit-Learn)

Balance Scale Data Set 

```python
import numpy as np
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
balance_data = pd.read_csv(
'https://archive.ics.uci.edu/ml/machine-learning-databases/balance-scale/balance-scale.data',
                           sep= ',', header= None)
```

```python
print("Dataset Lenght:: ", len(balance_data))
print("Dataset Shape:: ", balance_data.shape)
```

```python
print("Dataset:: ")
balance_data.head()
```

```python
X = balance_data.values[:, 1:5]
Y = balance_data.values[:,0]
```

```python
X_trainX , X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 100)
```

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_gini.fit(X_train, y_train)
```

```python
DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=3,
            max_features=None, max_leaf_nodes=None, min_samples_leaf=5,
            min_samples_split=2, min_weight_fraction_leaf=0.0,
            presort=False, random_state=100, splitter='best')
```

```python
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
 max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
DecisionTreeClassifier(class_weight=None, criterion='entropy', max_depth=3,
            max_features=None, max_leaf_nodes=None, min_samples_leaf=5,
            min_samples_split=2, min_weight_fraction_leaf=0.0,
            presort=False, random_state=100, splitter='best')
```

```python
clf_gini.predict([[4, 4, 3, 3]])
```

Prediction for Decision Tree classifier with criterion as gini index

```python
y_pred = clf_gini.predict(X_test)
y_pred
```

Prediction for Decision Tree classifier with criterion as information gain

```python
y_pred_en = clf_entropy.predict(X_test)
y_pred_en
```

Calculating Accuracy Score
The function accuracy_score() will be used to print accuracy of Decision Tree algorithm.
It takes 4 parameters.
y_true,
y_pred,
normalize,
sample_weight.

Accuracy for Decision Tree classifier with criterion as gini index

```python
print("Accuracy is ", accuracy_score(y_test,y_pred)*100)
```

Accuracy for Decision Tree classifier with criterion as information gain

```python
print("Accuracy is ", accuracy_score(y_test,y_pred_en)*100)
```
